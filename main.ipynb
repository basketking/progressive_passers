```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from adjustText import adjust_text
import os
import matplotlib.font_manager as fm
from pylab import rcParams
import requests
import lxml
import matplotlib as mpl
```

```python
## 1 - build links list for df_passing
## 1 - df_passing - (Passing table of players)

links_passing = [
    'https://fbref.com/en/comps/9/passing/Premier-League-Stats',
    'https://fbref.com/en/comps/12/passing/La-Liga-Stats',
    'https://fbref.com/en/comps/13/passing/Ligue-1-Stats',
    'https://fbref.com/en/comps/20/passing/Bundesliga-Stats',
    'https://fbref.com/en/comps/11/passing/Serie-A-Stats'
]
```

```python
df_passing = pd.DataFrame()
for link in links_passing:
    html_content = requests.get(link).text.replace('<!--', '').replace('-->', '')
    data = pd.read_html(html_content)[2]
    data['League'] = link.split('/')[-1][:-6]
    print(data.shape)
    df_passing = df_passing.append(data,ignore_index=True)

df_passing.shape # checking the output
```

```python
df_passing.columns = df_passing.columns.map('|'.join).str.strip('|')
```

```python
df_passing.columns
```

```python
df_passing = df_passing[['Unnamed: 1_level_0|Player',
       'Unnamed: 2_level_0|Nation', 'Unnamed: 3_level_0|Pos',
       'Unnamed: 4_level_0|Squad', 'Unnamed: 5_level_0|Age',
       'Unnamed: 7_level_0|90s', 'Total|Cmp',
       'Total|Att', 'Total|Cmp%', 'Total|TotDist', 'Total|PrgDist',
       'Unnamed: 26_level_0|1/3', 'League']]
```

```python
## Take out goalies from progressive passing

df_passing = df_passing[df_passing['Unnamed: 3_level_0|Pos'] != 'GK']

## Take out anomalies (where player name is player)

df_passing = df_passing[df_passing['Unnamed: 1_level_0|Player'] != 'Player']
```

```python
df_passing = df_passing[df_passing['Total|PrgDist'].notna()]
```

```python
## Cast strings to integers

df_passing['Total|PrgDist'] = df_passing['Total|PrgDist'].astype(int)
df_passing['Unnamed: 26_level_0|1/3'] = df_passing['Unnamed: 26_level_0|1/3'].astype(int)
```
